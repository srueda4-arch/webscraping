```python
# Navigating a Tree
# this is an html document, also known as a family tree of boxes 

<html>
<head>
    <title>The Dormouses story</title>
</head>
<body>
<h1 id="heading" class="page error"> Page 1 </h1>
<p class="title"><b>The Dormouses story</b></p>
<p class="story"> Once upon a time there were three little sisters; and their names were 
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a>
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>; 
and they lived at the bottom of a well. </p>

<p class="story">...</p>

</body> 
</html>
```

```python
# Importing the BeautifulSoup Library

# bs4 = the folder (package)
# BeautifulSoup = main tool inside that folder 

from bs4 import BeautifulSoup

# brings the BeautifulSoup class into Python so you can use it 
```

```python
# Reading the HTML file

# open = opens HTML file 
# .read() = reads the entire file as text 
# inp = stores raw HTML in string variable called inp 

fhand = open('Example 3.html','r')
inp = fhand.read()
```

```python
# Beautiful Soup Objects: python objects that are created when HTML is parsed 

# each object represents part of the webpage 
# converts HTML into python objects to work with 

# 3 Main Beautiful Soup Objects:
```

```python
# 1. BeautifulSoup Object: represents the entire webpage (whole HTML document)

# Creating a beautifulSoup object
# inp = string object, contains raw HTML 
# soup = BeautifulSoup object, contains parseed webpage tree 
soup = BeautifulSoup(inp, 'html.parser')
# take HTML text stored in variable inp, run it through BeautifulSoup using html.parser, store organized webpage into object called soup

# What is the class of the soup object?
print(type(soup))
# soup is a Beautiful Soup Object

# What is the name of the soup object?
print(soup.name)
# most objects in soup are tag objects, tags have names, but soup represents entire webpage = document 
```

```python
# 2. BeautifulSoup Tag Object
tag = soup.h1 # grabs h1 element, stores element inside python variable
# tag now represents = <h1 class="page error" id="heading"> Page 1 </h1>

# What is the class of the tag object?
print(type(tag))
# represents element class

# Print the tag
print(tag) # you could also code print(soup.h1)

# Access a tag name
print(tag.name) # you could also code print(soup.h1.name)
# name of html element is h1

# Access a tag's attributes - Finding the attribute value
print(tag['id'])
#  returns the value inside of the HTML attribute 

# Access a tag's attributes - Finding all attributes
print(tag.attrs)
# returns all attributes, class becomes a list ecause beautifulsoup stores it as ['page', 'error']
```

```python
# 2. BeautifulSoup Tag Object (alternative way)

# class: h1 is an HTML element  
print(type(soup.h1))

# print 
print(soup.h1)

# name: h1 is the name of the tag object
print(soup.h1.name)

# attributes: 
print(soup.h1['id'])

# all attributes
print(soup.h1.attrs)
```

```python
# 3. BeautifulSoup NavigableStrings Object
title_tag = soup.title
# grabs HTML with title tag <title> </title> and stores it in python variable title_tag

# What is the class?
print(type(title_tag.string))
# when you pull text out of a tag, doesn't return plain string, returns navigatable string 


# Extract the bit of text within a tag
print(title_tag.string)
# title_tag.string: give me the text only inside the tag 
```

```python
# 4. BeautifulSoup Comments Object
markup = "<b><!--Hey, buddy. Want to buy a used parser?--></b>"
soup1 = BeautifulSoup(markup, 'html.parser')
comment = soup1.b.string
```

```python
# Knowing the class
print(type(comment))
```

```python
# Printing the bit of string in the comment tag
print(comment)
```

```python
# Navigating the tree
# Going down the tree
# Navigating using Tag names
print(soup.head)
print(soup.title)
print(soup.body.b)
print(soup.a)
```

```python
# Navigating using .contents
head_tag = soup.head
print(head_tag)
print(head_tag.contents)
title_tag = head_tag.contents[1]
print("title tag", title_tag)
print(title_tag.contents)
```

```python
# Navigating using .children
for child in head_tag.children:
print(child)
```

```python
# Navigating using descendants
for child in head_tag.descendants:
print(child)
```

```python
# Navigating using string
print(title_tag.string)
print(head_tag.string)
```

```python
# Navigating using strings
for string in soup.strings:
print(repr(string))
```

```python
# Going up the tree
# Navigating using parent
title_tag = soup.title
print(title_tag)
print(title_tag.parent)
print(title_tag.string.parent)
html_tag = soup.html
print(type(html_tag.parent))
print(soup.parent)
```

```python
# Navigating using parents
link = soup.a
print(link)
for parent in link.parents:
if parent is None:
print(parent)
else:
print(parent.name)
```
