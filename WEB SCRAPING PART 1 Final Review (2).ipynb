```python
# Navigating a Tree
# this is an html document, also known as a family tree of boxes 

<html>
    <head>
        <title>The Dormouses story</title>
    </head>
    
<body>
    <h1 id="heading" class="page error"> Page 1 </h1>

    <p class="title"><b>The Dormouses story</b></p>
    <p class="story"> Once upon a time there were three little sisters; and their names were 
        <a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>
        <a href="http://example.com/lacie" class="sister" id="link2">Lacie</a>
        <a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>; 
        and they lived at the bottom of a well. </p>
    
    <p class="story">...</p>

</body> 
</html>
```

```python
# Importing the BeautifulSoup Library

# bs4 = the folder (package)
# BeautifulSoup = main tool inside that folder 

from bs4 import BeautifulSoup

# brings the BeautifulSoup class into Python so you can use it 
```

```python
# Reading the HTML file

# open = opens HTML file 
# .read() = reads the entire file as text 
# inp = stores raw HTML in string variable called inp 

fhand = open('Example 3.html','r')
inp = fhand.read()
```

```python
# Beautiful Soup Objects: python objects that are created when HTML is parsed 

# each object represents part of the webpage 
# converts HTML into python objects to work with 

# 3 Main Beautiful Soup Objects:
```

```python
# 1. BeautifulSoup Object: represents the entire webpage (whole HTML document)

# Creating a beautifulSoup object
# inp = string object, contains raw HTML 
# soup = BeautifulSoup object, contains parseed webpage tree 
soup = BeautifulSoup(inp, 'html.parser')
# take HTML text stored in variable inp, run it through BeautifulSoup using html.parser, store organized webpage into object called soup

# What is the class of the soup object?
print(type(soup))
# soup is a Beautiful Soup Object

# What is the name of the soup object?
print(soup.name)
# most objects in soup are tag objects, tags have names, but soup represents entire webpage = document 
```

```python
# 2. BeautifulSoup Tag Object

tag = soup.h1 # grabs h1 element, stores element inside python variable
# tag now represents = <h1 class="page error" id="heading"> Page 1 </h1>

# What is the class of the tag object?
print(type(tag))
# represents element class

# Print the tag: when you print a tag object, python displays actual HTML code for that tag 
print(tag) # you could also code print(soup.h1)

# Access a tag name
print(tag.name) # you could also code print(soup.h1.name)
# name of html element is h1

# Access a tag's attributes - Finding the attribute value
print(tag['id'])
#  returns the value inside of the HTML attribute 

# Access a tag's attributes - Finding all attributes
print(tag.attrs)
# returns all attributes, class becomes a list ecause beautifulsoup stores it as ['page', 'error']
```

```python
# 2. BeautifulSoup Tag Object (alternative way)

# class: h1 is an HTML (element / tag) 
print(type(soup.h1))

# print: when you print a tag object, python displays actual HTML code for that tag 
print(soup.h1)

# name: h1 is the name of the tag object
print(soup.h1.name)

# attributes: 
print(soup.h1['id'])

# all attributes
print(soup.h1.attrs)
```

```python
# 3. BeautifulSoup NavigableStrings Object

title_tag = soup.title
# grabs HTML with title tag <title> </title> and stores it in python variable title_tag

# What is the class?
print(type(title_tag.string))
# when you pull text out of a tag, doesn't return plain string, returns navigable string 
# title is an HTML (element/ NavigableString)

# Extract the bit of text within a tag
print(title_tag.string)
# title_tag.string: give me the text only inside the tag 
```

```python
# 4. BeautifulSoup Comments Object

markup = "<b><!--Hey, buddy. Want to buy a used parser?--></b>"
# takes a small HTML snippet
# <---...---> is an HTML comment 
# <b> is a bold tag

soup1 = BeautifulSoup(markup, 'html.parser')
# reads the HTML, creates the object tree 

comment = soup1.b.string
# soup1 grabs the b tag
# inside <b> is not normal text its an HTML comment embedded
# python does not return navigable string = plain text, it returns a comment object = embedded text

# Knowing the class
print(type(comment))
# class is and HTML (element/comment)

# Printing the bit of string in the comment tag
print(comment)
# ONLY INSIDE THE <---....--->
```

```python
# Navigating the tree
# Going down the tree
# Navigating using Tag names

# when you print a tag object, python displays actual HTML code for that tag 
print(soup.head)
print(soup.title)
print(soup.body.b) # gets body first, then gets the first b inside body 
print(soup.a) # gets the first a tag in page 
```

```python
# Navigating using .contents
head_tag = soup.head # grabs the whole head tag section, start to close and stores it in variable head_tag
print(head_tag) # prints entire head tag section 
print(head_tag.contents) # .contents returns a python list, returns what's inside the tag <head>, need /n = because that represents a new line 
title_tag = head_tag.contents[1] # list turns into indexes starting at 0
print("title tag", title_tag) # index 1 is <title> The Doormous story </title>
print(title_tag.contents) #.contents returns python list, and shows whats inside the tag <title>
```

```python
# Navigating using .children
# earlier we set head_tag=soup.head 
# give me only the immediate child of this tag 

for child in head_tag.children:
    print(child)
# YOU NEED THE TREE VIEW
        # head 
            # <title> 
                # "The Dormouse's story"
# .children = returns the only immediate child nested inside head, start to close 
# <title> The Dormouse story </title> is tag object 
```

```python
# Navigating using .descendants
# earlier we set head_tag=soup.head 

for child in head_tag.descendants:
    print(child)

# YOU NEED THE TREE VIEW
 #   <head>
  #      <title> 
    #      "The Dormouse's story"

# .descendants = returns everything nested inside the tag, at ALL LEVELS 
# this includes children, grandkids, all nested text 

# <title> The Dormouse story </title> is tag object 
# The Dormouse story is navigable string object (text)
    
# <title> tag and navigablestring prints seperately because they are two different python objects 
# TREE
# head
    # <title> > tag object
        # "story text" > navigablestring object 
```

```python
# Navigating using string
# .string: prints text only if the element contains exactly one text node and no nested tags, otherwise returns none

print(title_tag.string) # returns text, only one text node 
print(head_tag.string) # returns none, embedded child 
```

```python
# Navigating using strings
# .strings: gives all the pieces of text in the entire webpage, ONLY TEXT

for string in soup.strings: # walks through the whole HTML tree and pulls out every navigablestring object 
    print(repr(string))

# repr(): shows exact version of text including spaces and newlines, instead of invisible spaces you see '/n'
```

```python
# Going up the tree
# before navigating down, .contents, .children, .descendants 
# Navigating using parent

# parent-child tree - always make the tree to best understand 
# <html>
    # <head>
        # <title> 
            # "The Dormouse story"

title_tag = soup.title #.title prints where the title tag appears, start and close and text inside of it 
print(title_tag)

print(title_tag.parent) # .parent prints the parent tag of .title which is <head> and everything inside of it start and close

print(title_tag.string.parent) # find parent of text inside <title> parent is <title>, so prints evrything start and close

html_tag = soup.html # grabs the html tag and stores it inside html_tag

print(type(html_tag.parent)) # parent of HTML is soup and type is beautifulsoup

print(soup.parent) # there is no parent of soup (beautiful soup), so returns none 
```

```python
# Navigating using parents
link = soup.a # grabs first <a> tag on page 
print(link) # prints a tag and everything inside of it 
for parent in link.parents: # goes through every parent in the HTML and iterates each loop
    if parent is None:
        print(parent)
    else:
        print(parent.name) # prints the name of the parent, the biggest parent is [document]

# navigablestrings (text nodes) are the smallest children
```

```python
# Tree Structure 

[document]   ← biggest parent
   |
<html>
   |
<body>
   |
<p>
   |
<a>
   |
"The Dormouse's story"  ← smallest child

# only tags have children 
```
